```
from transformers import pipeline
```

```
generater = pipeline(
    "text-generation", 
    model = "abeja/gpt2-large-japanese"
)
# "日本で一番高い山は"に続くテキストを生成
outputs = generater("日本で一番高い山は")
print(outputs[0]["generated_text"])
```

```
import pandas as pd

fill_mask = pipeline(
    "fill-mask",
    model = "cl-tohoku/bert-base-japanese-v3"
)
masked_text = "日本の首都は[MASK] である"
outputs = fill_mask(masked_text)
display(pd.DataFrame(outputs[:3]))
```

```
masked_text = "今日の映画は刺激的で楽しかった。この映画は [MASK] 。"
outputs = fill_mask(masked_text)
display(pd.DataFrame(outputs[:3]))
```

```
t2t_generater = pipeline(
    "text2text-generation",
    model = "retrieva-jp/t5-large-long"
)
masked_text = "江戸幕府を開いたのは、<extra_id_0> である"
outputs = t2t_generater(masked_text, eos_token_id=32098)
print(outputs[0]["generated_text"])
```
