```python
from transformers import pipeline
```

```python
generater = pipeline(
    "text-generation", 
    model = "abeja/gpt2-large-japanese"
)
# "日本で一番高い山は"に続くテキストを生成
outputs = generater("日本で一番高い山は")
print(outputs[0]["generated_text"])
```
