```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib


K = 50  # 単語列の最大長
D = 64  # 埋め込みの次元

# 位置符号行列を初期化
pos_enc = np.empty((K, D))

for i in range(K):
    for k in range(D//2):
        theta = i / (10000 ** (2 * k / D))
        pos_enc[i, 2*k] = np.sin(theta)
        pos_enc[i, 2*k+1] = np.cos(theta)

```

```python
# 行列を画像で表示
im = plt.imshow(pos_enc)
plt.xlabel("dimension")
plt.ylabel("position")
plt.colorbar(im)
plt.show()
```

```python
# 位置符号同士の内積を計算
dot_matrix = np.matmul(pos_enc, pos_enc.T)
# 行列を画像表示
im = plt.imshow(dot_matrix, origin="lower")
plt.xlabel("dimension")
plt.ylabel("position")
plt.colorbar(im)
plt.show()
# -> 近いトークン同士のほうが意味的・文法的な関連度が高くなりやすいことを表現できている
```
